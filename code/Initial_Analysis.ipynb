```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
import nikescrapi as nk
```

```python
path = '/opt/airflow/data/preprocessed-data.csv'
if os.path.exists(path):
    df = pd.read_csv(path)
else:
    nikeAPI = nk.NikeScrAPI(max_pages=300, path=path)
    df = nikeAPI.getData()
```

```python
df = df.drop('Unnamed: 0',axis=1) # Drop the Unnamed column created when saving as csv.
```

```python
sns.countplot(data=df,x='subtitle')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data=df,x='title')
plt.xticks(rotation=90)
plt.show()
```

```python
df[df.duplicated()] # Check if there are duplicates
```

```python
def plot_null(df):
    percent_nul_per_col = (df.isnull().sum() / len(df)) * 100
    percent_nul_per_col[percent_nul_per_col != 0].transpose().sort_values().plot(kind='bar') # Only plot the null values
```

```python
plot_null(df)
```
